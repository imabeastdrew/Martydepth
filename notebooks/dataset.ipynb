```python
import gzip
import json

with gzip.open('data/raw/hooktheory (1).json.gz', 'r') as f:
    dataset = json.load(f)

print(len(dataset))
```

```python
# filter

train_set =  {
    k:v for k,v in dataset.items()
    if v['split'] == 'TRAIN'
    and 'AUDIO_AVAILABLE' in v['tags']
    and 'MELODY' in v['tags']
    and 'TEMPO_CHANGES' in v['tags']
}
print(len(train_set))
```
